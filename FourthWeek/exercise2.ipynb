```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
from scipy.stats import zscore

```

The following table shows the price distribution per liter of E10 for 20 petrol stations in the Aschaffenburg area:

```python
prices = np.repeat([1.70, 1.71, 1.73, 1.74, 1.75, 1.76, 1.79], repeats=[1, 4, 8, 2, 1, 1, 3])


prices
```

Determine the mean value

Determine the median

Determine the standard deviation

```python
mean_value = np.mean(prices)

median_value = np.median(prices)

std_dev = np.std(prices, ddof=1)
```

Draw a histogram with the breaks 1.70€, 1.72€, 1.74€, ...

```python
plt.figure(figsize=(10, 5))

plt.hist(prices, bins=np.arange(1.70, 1.81, 0.02), edgecolor='black')

plt.title("Price Distribution of E10 per Liter")

plt.xlabel("Price (€)")

plt.ylabel("Count")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
z_scores = zscore(prices)
z_scores
```

Standardize your data, i.e. determine the z-scores and display them in a histogram. What are the mean and standard deviation of the z-scores?

```python
plt.figure(figsize=(10, 5))

plt.hist(z_scores, bins=7, edgecolor='black')

plt.title("Histogram of Z-scores")

plt.xlabel("Z-score")

plt.ylabel("Count")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Mean and Standard Deviation of Z-scores

z_mean = np.mean(z_scores)

z_std_dev = np.std(z_scores, ddof=1)



(mean_value, median_value, std_dev, z_mean, z_std_dev)
```
